Issues:
- Age variable has two missing values
- Most players recorded almost no playtime while a few accumulated very high totals

- All demographic information is self-reported, so players may provide inaccurate ages or experience levels
- The response variable subscribe has many more TRUE than FALSE values
- hashedEmail is anonymized, duplicate players cannot be manually verified.
